```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing dataset for analysis
df = pd.read_csv(r'C:\Users\ho\Desktop\ifood_df.csv')
df.head()
```

```python
#deleting duplicate entries
df.drop_duplicates()
df.head()
```

```python
#checking null values
df.isna().sum()
```

```python
#checking unique entries
df.nunique()
```

```python
#z_costcontact and Z_Revenue these columns have only a single unique entry and may be not as usefull, Deleted these columns
df = df.drop(columns = ['Z_CostContact','Z_Revenue'])
```

```python
#descriptive stats of dataset
df.describe()
```

```python
#mode values
df.mode().dropna()
```

```python
#creating a column to show whether in a relation or not
def get_relationship(row):
    if row['marital_Married']==1:
        return 1
    elif row['marital_Together']==1:
        return 1
    else:
        return 0
df['in_relationship']= df.apply(get_relationship,axis=1)
df.head()
```

```python
#correcting a column name
df.rename(columns={'education_2n Cycle\t':'education_2n Cycle'}, inplace=True)
```

```python
#Assigning Education status level 1 to 5 in a different row

def Education_status(row):
    if row['education_2n Cycle'] == 1:
        return '2'
    elif row['education_Basic'] == 1:
        return '1'
    elif row['education_Graduation'] == 1:
        return '3'
    elif row['education_Master'] == 1:
        return '4'
    elif row['education_PhD'] == 1:
        return '5'
    else:
        return 'unknown'
df['Education_status'] = df.apply(Education_status, axis=1)   
```

SEGMENTATION

```python
#import kmeans
from sklearn.cluster import KMeans
```

```python
#Standardising the scale of the differnet columns to be used in segmentation

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_to_cluster = ['Income', 'MntTotal', 'in_relationship', 'Education_status']
cols_to_cluster
df_scaled = df.copy()
df_scaled[cols_to_cluster] = scaler.fit_transform(df[cols_to_cluster])
```

```python
#choosing a range for cluster values for elbow Method
x = df_scaled[cols_to_cluster]
inertia_list = []
for k in range(1,11):
    inertia = KMeans(n_clusters=k , n_init=10).fit(x).inertia_
    inertia_list.append(inertia)
```

```python
#Elbow method plot visualization

plt.figure(figsize=[7,5])
plt.plot(range(1,11), inertia_list)
plt.title("Inertia vs. Number of Clusters")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()

#4 appears to be the ideal numbere of clusters
```

```python
#Fittng desired number of clusters into our model and creating the segments
model = KMeans(n_clusters=4, random_state = 7)
model.fit(df_scaled[cols_to_cluster])
df_scaled['cluster'] = model.predict(df_scaled[cols_to_cluster])
```

```python
#Generating most relevant components of the clusters 
from sklearn import decomposition
pca = decomposition.PCA(n_components=2)
pca_res = pca.fit_transform(df_scaled[cols_to_cluster])
df_scaled['pc1'] = pca_res[:,0]
df_scaled['pc2'] = pca_res[:,1]
```

```python
#visualization of the clusters
plt.figure(figsize=[8,6])
sns.scatterplot(x='pc1', y='pc2', data=df_scaled, hue = 'cluster', palette='viridis',)
plt.title("Cluster")
plt.xlabel("Principal Component 1 (pc1)")
plt.ylabel("Principal Component 2 (pc2)")
plt.legend(title='Clusters')
```

PROPERTIES OF THE CLUSTERS

```python
#Number of customers per cluster
customer_num = df_scaled.groupby('cluster')['MntTotal'].count().reset_index()

plt.figure(figsize=[8,6])
sns.barplot(x='cluster', y='MntTotal', data=customer_num, palette='viridis',)
plt.title("Cluster sizes")
plt.xlabel("Cluster")
plt.ylabel("customer_num")
plt.show()
```

```python
#forming clusters without modifying the original values of columns used to form clusters 

scaler = StandardScaler()
cols_to_cluster = ['Income', 'MntTotal', 'in_relationship', 'Education_status']
df_clustered = df.copy()

model = KMeans(n_clusters=4, random_state = 7)
model.fit(df_clustered[cols_to_cluster])
df_clustered['cluster'] = model.predict(df_clustered[cols_to_cluster])
df_clustered.head()
```

```python
#Average Income of Each cluster
df_clustered['Avg_income'] = df_clustered.groupby('cluster')['Income'].mean()

plt.figure(figsize=[8,6])
sns.barplot(x='cluster', y='Avg_income', data=df_clustered, palette='viridis',)
plt.title("Cluster Avg_income")
plt.xlabel("Cluster")
plt.ylabel('Avg_income')
plt.show()
```

```python
#Average product consumption per cluster 
Avg_Consumption = df_clustered.groupby('cluster')[['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds','MntTotal']].mean().reset_index()
Avg_Consumption
```

```python
#Restructing the Data using Melt function
melted_dt = pd.melt(Avg_Consumption, id_vars='cluster', var_name="Product", value_name='Consumption')

plt.figure(figsize=[12,10])
sns.barplot(x='cluster', y='Consumption', data=melted_dt, errorbar=None, hue='Product')
plt.title("Product Consumption by Cluster")
plt.xlabel("Cluster")
plt.ylabel('Product consumption')
plt.xticks(rotation=0)
plt.legend(title="Product",loc="upper right")
plt.show()
```
