```python
df = df.drop_duplicates( keep='first')
df.duplicated().sum()
```

```python
# disreibution of legit transection and fradulent transection
df['Class'].value_counts()
```

```python
#this data set is higly unbalanced
#0 represents legit trans
#1 represents fraud trans
legit =df[df.Class == 0]
fraud =df[df.Class == 1]
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compares the values for both trans
df.groupby('Class').mean()
```

```python
#under sampling
legit_sample = legit.sample(n=492)
new_dataset = pd.concat([legit_sample, fraud], axis=0)
new_dataset.describe()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X1= new_dataset.drop(columns='Class', axis=1)
print(X1)
```

```python
y1 = new_dataset['Class']
print(y1)
```

```python
# MODEL 1
X1_train,X1_test,y1_train,y1_test=train_test_split(X1,y1,test_size=0.3,random_state=123)
print(X1.shape,X1_train.shape , X1_test.shape)
```

```python
lr=LogisticRegression()
model2=lr.fit(X1_train,y1_train)
prediction2=model2.predict(X1_test)
accuracy_score(y1_test,prediction2)
```

```python
# MODEL 2
from sklearn.model_selection import train_test_split
X2= new_dataset.drop(columns='Class', axis=1)
y2 = new_dataset['Class']
X2_train,X2_test,y2_train,y2_test=train_test_split(X2,y2,test_size=0.30,random_state=123)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
model=rfc.fit(X2_train,y2_train)
prediction=model.predict(X2_test)
accuracy_score(y2_test,prediction)
```

 Hence, Both of the models are performing with good accuracy
